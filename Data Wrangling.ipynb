```python
import geopandas as gpd
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
%pylab inline
```

KML to GeoJson:

```python
bins = gpd.GeoDataFrame.from_file('recyclingbins.json')
waterfacilities = gpd.GeoDataFrame.from_file('waterfacilities.json')
parks = gpd.GeoDataFrame.from_file('parks.json')
d_clusters = gpd.GeoDataFrame.from_file('dengue_clusters.json')
```

```python
print len(bins) # Number of bins in Singapore
print len(waterfacilities) # Number of facilities in Singapore
print len(parks) # Number of parks in Singapore
print len(d_clusters) # Number of cases in Singapore
```
